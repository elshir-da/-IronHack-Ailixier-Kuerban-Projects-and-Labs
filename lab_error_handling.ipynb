```python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input("Enter the quantity of " + product + "s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
```

```python
def calculate_total_price(customer_orders):
    total_price = 0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = int(input("Enter the price of " + product + ": "))
                if price >= 0:
                    total_price = total_price + price
                    valid_price = True
                else:
                    print("Price cannot be negative. Try again.")
            except ValueError:
                print("Invalid input. Please enter a valid integer price.")
    return total_price
```

```python
def get_customer_orders(inventory):
    valid_num = False
    while not valid_num:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders >= 0:
                valid_num = True
            else:
                print("Number of orders cannot be negative.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    customer_orders = set()

    for i in range(num_orders):
        valid_product = False
        while not valid_product:
            product_item = input("Enter the name of a product that a customer wants to order: ")
            
            if product_item in inventory and inventory[product_item] > 0:
                customer_orders.add(product_item)
                valid_product = True
            else:
                print("Error: Product is either out of stock or does not exist in inventory. Please try again.")
                
    return customer_orders
```

```python
products_list = ["t-shirt", "mug", "hat", "book", "keychain"]
```

```python
my_inventory_dict = initialize_inventory(products_list)
```

```python
my_orders_set = get_customer_orders(my_inventory_dict)
```

```python
final_total_price = calculate_total_price(my_orders_set)
print("Total Price:", final_total_price)
```
