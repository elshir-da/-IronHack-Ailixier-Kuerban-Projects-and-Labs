```python
import pandas as pd
import numpy as ny
```

```python
data_df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
```

```python
data_df.info()
```

```python
data_df.head()
```

```python
clean_df = data_df.copy()
```

```python
clean_df.columns = [column.lower().replace(" ","_") for column in clean_df.columns]
```

```python
clean_df = clean_df.rename(columns={"st":"state"})
```

```python
clean_df.head()
```

```python
clean_df["gender"].unique()
```

```python
standardized = {"F":"F",
                "M":"M",
                "Femal":"F",
                "Male":"M",
                "female":"F"}
```

```python
clean_df["gender"] = clean_df["gender"].replace(standardized)
```

```python
clean_df["gender"] = clean_df["gender"].fillna("Unknown")
```

```python
clean_df["state"].unique()
```

```python
clean_df["state"] = clean_df["state"].str.strip()
```

```python
state_name = {
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington",
    "NV": "Nevada"
}
```

```python
clean_df["state"] = clean_df["state"].replace(state_name)
```

```python
clean_df["state"] = clean_df["state"].fillna("Unknown")
```

```python
clean_df["state"].unique()
```

```python
clean_df.head()
```

```python
clean_df["education"].unique()
```

```python
clean_df["education"] = clean_df["education"].replace("Bachelors","Bachelor")
```

```python
clean_df["customer_lifetime_value"].unique()
```

```python
clean_df["customer_lifetime_value"] = clean_df["customer_lifetime_value"].str.replace("%","")
```

```python
clean_df["customer_lifetime_value"].unique()
```

```python
clean_df["vehicle_class"].unique()
```

```python
luxury = {"Luxury SUV":"Luxury",
          "Sports Car":"Luxury",
          "Luxury Car":"Luxury"}
```

```python
clean_df["vehicle_class"] = clean_df["vehicle_class"].replace(luxury)
```

```python
clean_df["vehicle_class"].unique()
```

```python
clean_df["customer_lifetime_value"] = clean_df["customer_lifetime_value"].astype(float)
clean_df["customer_lifetime_value"].unique()
```

```python
clean_df["number_of_open_complaints"].unique()
```

```python
clean_df["number_of_open_complaints"] = clean_df["number_of_open_complaints"].fillna("0/0/0")
clean_df["number_of_open_complaints"].unique()
```

```python
clean_df["number_of_open_complaints"] = clean_df["number_of_open_complaints"].astype(str)
```

```python
split_data = clean_df["number_of_open_complaints"].str.split("/")
```

```python
clean_df["number_of_open_complaints"] = split_data.str[1]
```

```python
clean_df["number_of_open_complaints"] = pd.to_numeric(clean_df["number_of_open_complaints"], errors='coerce')
```

```python
clean_df["number_of_open_complaints"].unique()
```

```python
null_counts = clean_df.isnull().sum()
print(null_counts[null_counts > 0])
```

```python

clean_df["customer_lifetime_value"] = clean_df["customer_lifetime_value"].fillna(0)
clean_df["income"] = clean_df["income"].fillna(0)
clean_df["monthly_premium_auto"] = clean_df["monthly_premium_auto"].fillna(0)
clean_df["total_claim_amount"] = clean_df["total_claim_amount"].fillna(0)


clean_df["customer"] = clean_df["customer"].fillna("Unknown")
clean_df["education"] = clean_df["education"].fillna("Unknown")
clean_df["policy_type"] = clean_df["policy_type"].fillna("Unknown")
clean_df["vehicle_class"] = clean_df["vehicle_class"].fillna("Unknown")


clean_df.isnull().sum()
```

```python

clean_df["customer_lifetime_value"] = clean_df["customer_lifetime_value"].astype(int)
clean_df["income"] = clean_df["income"].astype(int)
clean_df["monthly_premium_auto"] = clean_df["monthly_premium_auto"].astype(int)
clean_df["total_claim_amount"] = clean_df["total_claim_amount"].astype(int)

clean_df["number_of_open_complaints"] = clean_df["number_of_open_complaints"].astype(int)


clean_df.info()
```

```python
clean_df.duplicated().sum()
```

```python
clean_df = clean_df.drop_duplicates(keep="first")


clean_df = clean_df.reset_index(drop=True)


print(clean_df.duplicated().sum())
```

```python
clean_df.to_csv("clean_data.csv", index=False)
```
