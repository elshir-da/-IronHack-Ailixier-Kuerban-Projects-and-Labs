```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
```

```python
def initialize_inventory(products):
    inventory = {product: int(input("Enter the quantity for " + product + ": ")) for product in products}
    return inventory
```

```python
def get_customer_orders():
    num_orders = int(input("Enter the number of customer orders: "))

    customer_orders = {input("Enter the name of a product that a customer wants to order: ") for i in range(num_orders)}
    return customer_orders
```

```python
def calculate_total_price(customer_orders):
    total_price = sum(int(input("Enter the price of " + product + ": ")) for product in customer_orders)
    return total_price
```

```python
def update_inventory(customer_orders, inventory):
    for order in customer_orders:
        if order in inventory:
            inventory[order] = inventory[order] - 1


    new_update_inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return new_update_inventory
```

```python
def print_order_statistics(customer_orders, products):
    total = len(customer_orders)
    percentage = (total / len(products)) + 100
    print("Order statistics: ")
    print("Total percentage ordered: ")
    print("Percentage of unique product ordered ", percentage, "%")
```

```python
def print_update_inventory(inventory):
    print("Update inventory:")
    for product in inventory:
        print(product + ":", inventory[product])
```

```python
my_inventory = initialize_inventory(products)
```

```python
my_orders = get_customer_orders()
```

```python
print_order_statistics(my_orders, products)
```

```python
my_inventory = update_inventory(my_orders, my_inventory)
```

```python
print_update_inventory(my_inventory)
```

```python
final_price = calculate_total_price(my_orders)
print("Total price:", final_price)
```
